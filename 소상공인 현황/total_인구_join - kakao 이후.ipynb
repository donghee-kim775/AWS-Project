```python
import pandas as pd

total_data = pd.read_csv('merged_data_완성본.csv', encoding='cp949', index_col=0)
cluster_data = pd.read_excel('cluster_연령분포.xlsx', index_col=0)
```

```python
total_data
selected_columns = ['읍면동', '도로명']
total_data = total_data.dropna(subset=selected_columns)
```

```python
#경산시
def convert_kyungsan_dong(row):
    dong = str(row['읍면동'])  # Convert to string
    if dong.endswith("동"):
        if dong == "삼북동":
            return "중앙동"
        elif dong in ["남방동", "내방동", "내동", "여천동", "신천동", "점촌동", "평산동", "사동", "삼풍동", "계양동"]:
            return "동부동"
        elif dong in ["옥곡동", "사정동", "옥산동"]:
            return "서부 1동"
        elif dong in ["중산동", "정평동"]:
            return "서부 2동"
        elif dong in ["삼남동", "서상동", "신교동", "상방동", "백천동"]:
            return "남부동"
        elif dong in ["대평동", "대정동", "임당동", "대동", "조영동", "갑제동", "계양동"]:
            return "북부동"
        elif dong == "중방동":
            return "중방동"
        else:
            return dong
    else:
        return dong

#경주시
def convert_kyungju_dong(row):
    dong = str(row['읍면동'])
    if dong.endswith("동"):
        if dong in ["동부동", "서부동", "북부동", "노동동", "노서동"]:
            return "중부동"
        elif dong in ["성동동", "황오동"]:
            return "황오동"
        elif dong in ["성건동", "석장동"]:
            return "성건동"
        elif dong in ["황남동", "사정동", "탑동", "배동", "율동"]:
            return "황남동"
        elif dong in ["인왕동", "교동", "동방동", "도지동", "남산동", "구황동", "보문동", "배반동"]:
            return "월성동"
        elif dong in ["충효동", "서악동", "효현동", "광명동"]:
            return "선도동"
        elif dong in ["진현동", "마동", "하동", "평동", "시래동", "구정동", "조양동", "시동"]:
            return "불국동"
        elif dong in ["신평동", "손곡동", "북군동", "천군동", "덕동", "황용동", "암곡동"]:
            return "보덕동"
        else:
            return dong
    else:
        return dong

#구미시
def convert_gumi_dong(row):
    dong = str(row['읍면동'])
    if dong.endswith("동"):
        if dong in ["지산동", "양호동"]:
            return "지산동"
        elif dong in ["봉곡동", "부곡동", "선기동", "수점동", "남통동"]:
            return "선주원남동"
        elif dong in ["광평동", "신평동", "송정동"]:
            return "광평동"
        elif dong in ["사곡동", "상모동", "임은동"]:
            return "상모사곡동"
        elif dong in ["임은동", "오태동"]:
            return "임오동"
        elif dong in ["인의동", "황상동", "신동", "구평동"]:
            return "인동동"
        elif dong in ["진평동", "시미동", "임수동"]:
            return "진미동"
        elif dong in ["구포동", "금전동", "양호동", "거의동", "옥계동"]:
            return "양포동"
        elif dong in ["형곡동"]:
            return "형곡1,2동"  #인원데이터에서 형곡1동, 형곡2동 합쳐서 형곡1,2동으로 네이밍
        elif dong in ["신평동"]:
            return "신평1동"
        elif dong in ["원평동", "광평동", "공단동"]:
            return "신평2동"
        else:
            return dong
    else:
        return dong

#김천시
def convert_kimchun_dong(row):
    dong = str(row['읍면동'])
    if dong.endswith("동"):
        if dong in ["감호동", "용두동", "모암동", "성내동"]:
            return "자산동"
        elif dong in ["평화동", "남산동"]:
            return "평화남산동"
        elif dong in ["황금동", "양천동"]:
            return "양금동"
        elif dong in ["신음동", "대광동", "응명동", "교동", "삼락동", "문당동"]:
            return "대신동"
        elif dong in ["다수동", "백옥동", "부곡동"]:
            return "대곡동"
        elif dong in ["지좌동", "덕곡동"]:
            return "지좌동"
        else:
            return dong
    else:
        return dong
    
#문경시 점천동 보류    
def convert_munkyung_dong(row):
    dong = str(row['읍면동'])
    if dong.endswith("동"):
        if dong in ["점촌동"]:
            return "점촌1동"
        elif dong in ["영신동", "윤직동"]:
            return "점촌2동"
        elif dong in ["흥덕동", "우지동", "창동"]:
            return "점촌3동"
        elif dong in ["공평동", "유곡동", "불정동", "신기동"]:
            return "점촌4동"
        elif dong in ["모전동"]:
            return "점촌5동"
        else:
            return dong
    else:
        return dong
        
#상주시
def convert_sangju_dong(row):
    dong = str(row['읍면동'])
    if dong.endswith("동"):
        if dong in ["남성동", "낙양동", "개운동", "연원동", "남장동"]:
            return "남원동"
        elif dong in ["무양동", "초산동", "부원동", "죽전동", "만산동", "남적동"]:
            return "북문동"
        elif dong in ["냉림동", "낙상동", "중덕동", "화산동", "계산동"]:
            return "계림동"
        elif dong in ["인봉동", "복룡동", "서성동", "서문동", "화개동", "외답동", "헌신동", "병성동", "도남동"]:
            return "동문동"
        elif dong in ["성하동", "성동동", "거동동", "인평동", "서곡동"]:
            return "동성동"
        elif dong in ["신봉동", "가장동", "양촌동", "지천동", "오대동", "흥각동"]:
            return "신흥동"
        else:
            return dong
    else:
        return dong
    
#안동시    
def convert_ahndong_dong(row):
    dong = str(row['읍면동'])
    if dong.endswith("동"):
        if dong in ["북문동", "삼산동", "서부동", "목성동", "동문동", "동부동", "운흥동", "남문동", "남부동", "천리동", "율세동", "옥정동", "신세동", "법흥동", "상아동"]:
            return "중구동"
        elif dong in ["명륜동", "신안동", "안막동", "상아동"]:
            return "명륜동"
        elif dong in ["용상동", "성곡동", "석동동", "송천동"]:
            return "용상동"
        elif dong in ["안흥동", "대석동", "옥야동", "광석동", "당북동", "법상동", "금곡동", "화성동"]:
            return "서구동"
        elif dong in ["평화동", "운안동"]:
            return "평화동"
        elif dong in ["안기동", "이천동"]:
            return "안기동"
        elif dong in ["송현동", "노하동"]:
            return "송하동"
        elif dong in ["정상동", "정하동", "수상동", "수하동"]:
            return "강남동"
        else:
            return dong
    else:
        return dong
    
    
#영주시 중복 보류
def convert_youngju_dong(row):
    dong = str(row['읍면동'])
    if dong.endswith("동"):
        if dong in ["상망동, 조와동"]:
            return "상망동"
        elif dong in ["휴천동"]:
            return "휴천1,2동"  # 인원데이터에서 휴천1동, 휴천2동 합쳐서 휴천1,2동으로 만들어야함
        elif dong in ["영주동"]:
            return "영주1,2동"   #영주1동, 영주2동 합쳐서 영주1,2동으로 만들어야함
        elif dong in ["안흥동", "대석동", "옥야동", "광석동", "당북동", "법상동", "금곡동", "화성동"]:
            return "서구동"
        elif dong in ["조암동", "적서동"]:
            return "휴천3동"
        elif dong in ["가흥동"]:
            return "가흥1동"
        elif dong in ["문정동", "고현동", "창진동", "상줄동", "아지동"]:
            return "가흥2동"
        else:
            return dong
    else:
        return dong
#영천시
def convert_youngchun_dong(row):
    dong = str(row['읍면동'])
    if dong.endswith("동"):
        if dong in ["조교동", "망정동", "야사동", "언하동", "신기동"]:
            return "동부동"
        elif dong in ["문내동", "문외동", "창구동", "과전동", "오미동", "녹전동", "도림동", "매산동"]:
            return "중앙동"
        elif dong in ["성내동", "화룡동", "교촌동", "오수동", "쌍계동", "대전동", "서산동"]:
            return "서부동"
        elif dong in ["도동", "금노동", "범어동", "작산동", "봉동", "도남동", "본촌동", "채신동", "괴연동"]:
            return "남부동"
        else:
            return dong
    else:
        return dong


    
#포항시 북구
def convert_pohangbukgu_dong(row):
    dong = str(row['읍면동'])
    if dong.endswith("동"):
        if dong in ["득량동", "학잠동"]:
            return "양학동"
        elif dong in ["우현동", "창포동"]:
            return "우창동"
        elif dong in ["중앙동","동빈동1가","동빈동2가","덕산동","덕수동","여천동","상원동","남빈동","신흥동","대흥동","항구동","대신동","학산동"]:
            return "중앙동"
        elif dong in ["장성동", "양덕동"]:
            return "장량동"
        elif dong in ["환호동", "여남동"]:
            return "환여동"
        else:
            return dong
    else:
        return dong
    
    
    
#포항시 남구
def convert_pohangnamgu_dong(row):
    dong = str(row['읍면동'])
    if dong.endswith("동"):
        if dong in ["대잠동", "이동"]:
            return "대이동"
        elif dong in ["상도동", "대도동"]:
            return "상대동"
        elif dong in ["송내동", "송정동", "괴동동", "동촌동", "장흥동", "인덕동", "호동"]:
            return "제철동"
        elif dong in ["청림동", "일월동"]:
            return "청림동"
        elif dong in ["효자동", "지곡동"]:
            return "효곡동"
        else:
            return dong
    else:
        return dong
    
convert_kyungsan_dong, convert_kyungju_dong, convert_gumi_dong, convert_kimchun_dong, convert_sangju_dong, convert_ahndong_dong, convert_youngchun_dong, convert_pohangbukgu_dong, convert_pohangnamgu_dong

# '시군구'가 '경산시' -> '읍면동'
total_data.loc[total_data['시군구'] == '경산시', '읍면동'] = total_data[total_data['시군구'] == '경산시'].apply(convert_kyungsan_dong, axis=1)
# '시군구'가 '경주시' -> '읍면동'
total_data.loc[total_data['시군구'] == '경주시', '읍면동'] = total_data[total_data['시군구'] == '경주시'].apply(convert_kyungju_dong, axis=1)
# '시군구'가 '구미시' -> '읍면동'
total_data.loc[total_data['시군구'] == '구미시', '읍면동'] = total_data[total_data['시군구'] == '구미시'].apply(convert_gumi_dong, axis=1)
# '시군구'가 '김천시' -> '읍면동'
total_data.loc[total_data['시군구'] == '김천시', '읍면동'] = total_data[total_data['시군구'] == '김천시'].apply(convert_kimchun_dong, axis=1)
# '시군구'가 '상주시' -> '읍면동'
total_data.loc[total_data['시군구'] == '상주시', '읍면동'] = total_data[total_data['시군구'] == '상주시'].apply(convert_sangju_dong, axis=1)
# '시군구'가 '안동시' -> '읍면동'
total_data.loc[total_data['시군구'] == '안동시', '읍면동'] = total_data[total_data['시군구'] == '안동시'].apply(convert_ahndong_dong, axis=1)
# '시군구'가 '영천시' -> '읍면동'
total_data.loc[total_data['시군구'] == '영천시', '읍면동'] = total_data[total_data['시군구'] == '영천시'].apply(convert_youngchun_dong, axis=1)
# '시군구'가 '포항북구시' -> '읍면동'
total_data.loc[total_data['시군구'] == '포항시 북구', '읍면동'] = total_data[total_data['시군구'] == '포항시 북구'].apply(convert_pohangbukgu_dong, axis=1)
# '시군구'가 '포항남구시' -> '읍면동'
total_data.loc[total_data['시군구'] == '포항시 남구', '읍면동'] = total_data[total_data['시군구'] == '포항시 남구'].apply(convert_pohangnamgu_dong, axis=1)
# '시군구'가 '문경시' -> '읍면동'
total_data.loc[total_data['시군구'] == '문경시', '읍면동'] = total_data[total_data['시군구'] == '문경시'].apply(convert_munkyung_dong, axis=1)
# '시군구'가 '영주시' -> '읍면동'
total_data.loc[total_data['시군구'] == '영주시', '읍면동'] = total_data[total_data['시군구'] == '영주시'].apply(convert_youngju_dong, axis=1)

total_data['시군구'] = total_data['시군구'].replace({"포항시북구" : "포항시 북구", "포항시남구" : "포항시 남구"})
```
