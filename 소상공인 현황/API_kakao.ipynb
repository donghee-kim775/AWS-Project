```python
import pandas as pd
import requests
import openpyxl as xl
import numpy as np
df1 = pd.read_csv('totaldata.csv', encoding='cp949')

df1
```

```python
addrs = df1['roadNmAddr']
addrs
```

```python
addrs = df1['시'] + ' ' + np.where(df1['구'].isnull(),'',df1['구']) + ' ' + df1['도로명']
addrs = addrs.unique()
len(addrs)

locations = []

for addr in addrs:
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query={}'.format(addr)
    headers = {
    ## 여러분의 카카오 API의 REST API키를 아래 예시와 같이 입력해주세요
    ## "Authorization": "KakaoAK REST API키 입력 gogo"}
    "Authorization": "KakaoAK b56dac02d90902d0da8612b99dfb431d"}
    place = requests.get(url, headers = headers).json()['documents']
    locations.append(place)
    print(addr)
```

```python
city = [] ## 시, 군
town = [] ## 동, 읍, 면
road = [] ## 해당 길
for i in range(len(locations)):
    try:
        # Extract information from the address
        address_info = locations[i][0].get('address', {})
        road_info = locations[i][0].get('road_address', {})

        # Get city, town, and road names
        city_name = address_info.get('region_2depth_name', '없음')
        town_name = address_info.get('region_3depth_name', '없음')
        road_name = road_info.get('road_name', '없음')

        # Append the values to the lists
        city.append(city_name)
        town.append(town_name)
        road.append(road_name)

    except IndexError:
        print(f"{i}번째 주소 못가져옴")
        city.append('없음')
        town.append('없음')
        road.append('없음')

    except AttributeError:
        print(f"{i}번째 주소 속성 오류")
        city.append('없음')
        town.append('없음')
        road.append('없음')
```

```python
city_town = np.array([city,town,road]).T
df_temp = pd.DataFrame(city_town, columns = ['region_2depth_name','region_3depth_name', 'road_name'])
df_temp

df_temp = df_temp[~(df_temp == '없음').any(axis=1)]
df_temp = df_temp.sort_values(by=['region_2depth_name', 'region_3depth_name', 'road_name'], ascending=[True, True, True])

df_temp
df_temp.to_excel('지번.xlsx')
```
