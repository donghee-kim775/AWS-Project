```python
import requests
import pandas as pd
import xml.etree.ElementTree as ET
encoding_key = 'tFlRbdLrGdbT07si0ZGVgNB2t7e1lRlQPzeq5uCeXyfkCZJ%2B8%2BCRvlRy1JLFEZXUdVqNxgmO1OaOcSDStjOhKQ%3D%3D'
```

```python
def xmltodf(contents):
    root = ET.fromstring(contents)
    items = root.findall(".//item")
    
    data = []
    for item in items:
        item_data = {}
        for child in item:
            item_data[child.tag] = child.text
        data.append(item_data)
        
    df = pd.DataFrame(data)
    return df
```

```python
def brand_api(encoding_key, pageNo, numRow, year):
    url = 'https://apis.data.go.kr/1130000/FftcBrandFrcsStatsService/getBrandFrcsStats?'
    serviceKey = 'serviceKey=' + encoding_key
    page = '&pageNo=' + str(pageNo)
    numOfRows = '&numOfRows=' + str(numRow)
    resulttype = '&resultType=xml'
    yr = '&yr=' + str(year)
    
    total_url = url + serviceKey + page + numOfRows + yr

    response = requests.get(total_url)
    contents = response.text
    df = xmltodf(contents)
    return df

df = brand_api(encoding_key, 1, 1000, 2019)

df = df.rename(columns = {'yr' : '기준년도', 'indutyLclasNm' : '업종대분류', 'indutyMlsfcNm' : '업종중분류', 'corpNm' : '법인명', 'brandNm' : '브랜드명'})
df = df.rename(columns = {'frcsCnt' : '가맹점수', 'newFrcsRgsCnt' : '신규가맹점등록수', 'ctrtEndCnt' : '계약종료수', 'ctrtCncltnCnt' : '계약해지수', 'nmChgCnt' : '명변경수'})
df = df.rename(columns = {'avrgSlsAmt' : '평균매출금액', 'arUnitAvrgSlsAmt' : '면적단위평균매출금액'})
df
```
