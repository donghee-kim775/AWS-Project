```python
import requests
import pandas as pd
import xml.etree.ElementTree as ET

decoding_key = 'tFlRbdLrGdbT07si0ZGVgNB2t7e1lRlQPzeq5uCeXyfkCZJ+8+CRvlRy1JLFEZXUdVqNxgmO1OaOcSDStjOhKQ=='
encoding_key = 'tFlRbdLrGdbT07si0ZGVgNB2t7e1lRlQPzeq5uCeXyfkCZJ%2B8%2BCRvlRy1JLFEZXUdVqNxgmO1OaOcSDStjOhKQ%3D%3D'

# XML 데이터 파싱
def xmltodf(xml_data):
    root = ET.fromstring(xml_data)
    items = root.findall(".//item")
    
    data = []
    for item in items:
        item_data = {}
        for child in item:
            item_data[child.tag] = child.text
        data.append(item_data)
        
    df = pd.DataFrame(data)
    
    return df
```

# 외식별 창업 비용 현황

```python
def api_Outcosts(encoding_key, pageNo, numRow, year):
    url = 'https://apis.data.go.kr/1130000/FftcSclasIndutyFntnStatsService/getSclaIndutyFntnOutStats?'
    serviceKey = 'serviceKey=' + encoding_key
    page = '&pageNo=' + str(pageNo)
    numofRows = '&numOfRows=' + str(numRow)
    result_type = '&resultType=xml'
    yr = '&yr=' + str(year)

    total_url = url + serviceKey + page + numofRows + result_type + yr

    response = requests.get(total_url)
    contents = response.text

    dataframe = xmltodf(contents)
    return dataframe

# for i in range(2019, 2023):
data = api_Outcosts(encoding_key, 1, 1000, 2022)
data = data.rename(columns = {'yr' : '기준년도', 'indutyMlsfcNm' : '업종중분류'})
data = data.rename(columns = {'jnghdqrtrsCnt' : '가맹본부수', 'frcsCnt' : '가맹점수'})
data = data.rename(columns = {'avrgFrcsAmt' : '평균가맹교육금액', 'avrgFntnAmt' : '평균가맹보증금액', 'avrgJngEtcAmt' : '평균가맹기타금액', 'smtnAmt' : '합계금액'})
data = data.rename(columns = {'crrncyUnitCdNm' : '화폐단위코드명'})
data

data
```
