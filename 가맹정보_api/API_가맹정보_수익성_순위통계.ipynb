```python
import requests
import pandas as pd
import xml.etree.ElementTree as ET
encoding_key = 'tFlRbdLrGdbT07si0ZGVgNB2t7e1lRlQPzeq5uCeXyfkCZJ%2B8%2BCRvlRy1JLFEZXUdVqNxgmO1OaOcSDStjOhKQ%3D%3D'
```

```python
def xmltodf(contents):
    root = ET.fromstring(contents)
    items = root.findall(".//item")
    
    data = []
    for item in items:
        item_data = {}
        for child in item:
            item_data[child.tag] = child.text
        data.append(item_data)
        
    df = pd.DataFrame(data)
    return df
```

# 가맹본부 수익성 순위 통계 조회

```python
def api1(encoding_key, pageNo, numRow, year):
    url = 'https://apis.data.go.kr/1130000/FftcJnghdqrtrsProfStatsService/getJnghdqrtrsProfStats?'
    serviceKey = 'serviceKey=' + encoding_key
    page = '&pageNo=' + str(pageNo)
    numofRows = '&numOfRows=' + str(numRow)
    yr = '&yr=' + str(year)

    total_url = url + serviceKey + page + numofRows + yr

    response = requests.get(total_url)
    contents = response.text
    df = xmltodf(contents)
    return df

for i in range(2019, 2023):
    df_1 = api1(encoding_key, 1, 1000, i)
    df_1 = df_1.rename(columns = {'yr' : '기준년도', 'indutyLclasNm' : '업종분류'})
    df_1 = df_1.rename(columns = {'thstrmNtpfAmt' : '자본금액', 'caplAmt' : '매출금액', 'slsAmt' : '영업이익금액', 'bsnProfitAmt' : '당기순이익금액'})
    df_1 = df_1.rename(columns = {'bsnProfitRt' : '영업이익율', 'slsNtpfRt' : '매출순이익율', 'ecptlNtpfRt' : '자기자본순이익율'})
df_1
```
