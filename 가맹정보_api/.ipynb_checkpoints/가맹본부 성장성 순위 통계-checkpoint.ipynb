```python
import requests
import pandas as pd
import xml.etree.ElementTree as ET
encoding_key = 'tFlRbdLrGdbT07si0ZGVgNB2t7e1lRlQPzeq5uCeXyfkCZJ%2B8%2BCRvlRy1JLFEZXUdVqNxgmO1OaOcSDStjOhKQ%3D%3D'
```

```python
def xmltodf(contents):
    root = ET.fromstring(contents)
    items = root.findall(".//item")
    
    data = []
    for item in items:
        item_data = {}
        for child in item:
            item_data[child.tag] = child.text
        data.append(item_data)
        
    df = pd.DataFrame(data)
    return df
```

# 가맹본부 성장성 순위 통계 조사

```python
def api1(encoding_key, pageNo, numRow, year):
    url = 'https://apis.data.go.kr/1130000/FftcJnghdqrtrsGrStatsService/getJnghdqrtrsGrStats?'
    serviceKey = '&serviceKey=' + encoding_key
    page = 'pageNo=' + str(pageNo)
    numofRows = '&numOfRows=' + str(numRow)
    resultType = '&resultType=xml'
    yr = '&yr=' + str(year)

    total_url = url + page + numofRows + serviceKey + resultType + yr

    response = requests.get(total_url)
    contents = response.text
    df = xmltodf(contents)
    return df
```

```python
df = api1(encoding_key, 1, 1000, 2019)
df = df.rename(columns = {'yr' : '기준년도', 'indutyLclasNm' : '업종대분류명'})
df = df.rename(columns = {'assetsAmt' : '자산금액', 'slsAmt' : '영업이익금액', 'bsnProfitAmt' : '영업이익금액'})
df = df.rename(columns = {'assetsIncRt' : '자산증가율', 'slsIncRt' : '매출증가율', 'bsnProfitIncRt' : '영업이익증가율'})
df
```
