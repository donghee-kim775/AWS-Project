# 브랜드 업종별 가맹점 분포 통계 조회

```python
#함수정의
def API_Affiliate(decoding_key, pageNo, numRows, year):
    url = 'https://apis.data.go.kr/1130000/FftcBrandIndutyDropFrcsStatsService/getBrandIndutyFrcsStats'
    service_key = '?serviceKey=' + decoding_key
    page = '&pageNo=' + str(pageNo)
    numofRows = '&numOfRows=' + str(numRows)
    yr = '&resultType=json&yr=' + str(year)
    
    total_url = url + service_key + page + numofRows + yr
    
    response = requests.get(total_url)
    contents = response.text

    # json load
    json_data = json.loads(contents)
    body = json_data['items']
    df = pd.DataFrame(body)
    df.rename(columns={'yr':'기준년도','indutyLclasNm':'업종대분류명','frcsSeNm':'가맹점구분명','frcsSeBrandCnt':'가맹점구분별 브랜드수','indutyJngSeBrandCnt':'업종별가맹점구분별 브랜드수','indutyLclasBrandCntRt':'업종별가맹점브랜드비율'}, inplace=True)
    df
    return df

# API
API_Affiliate(decoding_key, 1, 1000, 2019)

# 년도별 API 불러오기
# for i in range(2017, 2023):
#    a = API_Affiliate(decoding_key, 1, 1000, i)       
#a
```
