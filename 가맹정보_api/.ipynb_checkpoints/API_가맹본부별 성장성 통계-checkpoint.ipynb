```python
import pandas as pd
import requests
api_key = '/SP+KE9l4FB+wWeiL0Xg7Awr43/tw6K4EM62dzSju74IU3fnqry+ajNhq7aIEaoFKmHs35SYMtQ2lzOyl3VudA=='
base_url = 'https://apis.data.go.kr/1130000/FftcJnghdqrtrsGrStatsService/getJnghdqrtrsGrStats'

params = {
    'serviceKey': api_key,
    'pageNo': 1,
    'numOfRows': 10,
    'resultType': 'json',
    'yr': 2022,
}

response = requests.get(base_url, params=params)
data = response.json()

df = pd.DataFrame(data['items'])

# 필요한 컬럼만 선택하여 새로운 DataFrame 생성
selected_columns = ['yr', 'indutyLclasNm', 'assetsAmt', 'assetsIncRt']
df = df[selected_columns]
df = df.rename(columns={
    'yr': '연도',
    'indutyLclasNm': '산업 대분류 이름',
    'assetsAmt': '자산 총액',
    'assetsIncRt': '자산 증가율'
})

df
```
