```python
import requests
import pandas as pd
import xml.etree.ElementTree as ET

decoding_key = 'tFlRbdLrGdbT07si0ZGVgNB2t7e1lRlQPzeq5uCeXyfkCZJ+8+CRvlRy1JLFEZXUdVqNxgmO1OaOcSDStjOhKQ=='
encoding_key = 'tFlRbdLrGdbT07si0ZGVgNB2t7e1lRlQPzeq5uCeXyfkCZJ%2B8%2BCRvlRy1JLFEZXUdVqNxgmO1OaOcSDStjOhKQ%3D%3D'

# XML 데이터 파싱
def xmltodf(xml_data):
    root = ET.fromstring(xml_data)
    items = root.findall(".//item")
    
    data = []
    for item in items:
        item_data = {}
        for child in item:
            item_data[child.tag] = child.text
        data.append(item_data)
        
    df = pd.DataFrame(data)
    
    return df
```

```python
def brand_costs(encoding_key, pageNo, numRow, year):
    url = 'https://apis.data.go.kr/1130000/FftcBrandFntnStatsService/getBrandFntnStats?'
    serviceKey = 'serviceKey=' + encoding_key
    page = '&pageNo=' + str(pageNo)
    numOfRows = '&numOfRows=' + str(numRow)
    result_type = '&resultType=xml'
    yr = '&yr=' + str(year)

    total_url = url + serviceKey + page + numOfRows + result_type + yr

    response = requests.get(total_url)
    contents = response.text

    dataframe = xmltodf(contents)
    return dataframe

data = brand_costs(encoding_key, 1, 1000, 2017)
data = data.rename(columns = {'yr' : '기준년도', 'indutyLclasNm' : '업종대분류명', 'indutyMlsfcNm' : '업종준분류명'})
data = data.rename(columns = {'brandNm' : '브랜드명', 'corpNm' : '법인명'})
data = data.rename(columns = {'jngBzmnJngAmt' : '가맹사업자가맹금액', 'jngBzmnEduAmt' : '사업자교육금액'})
data = data.rename(columns = {'jngBzmnAssrncAmt' : '가맹사업자보증금액', 'jngBzmnEtcAmt' : '가맹사업자기타금액', 'smtnAmt' : '합계금액'})
data
```
